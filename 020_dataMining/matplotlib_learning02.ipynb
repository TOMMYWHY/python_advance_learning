### demo03

```python
fandango =  pd.read_csv("fandango_score_comparison.csv") # movies
num_cols = ["FILM","Metacritic_User","IMDB","Metacritic_norm","Fandango_Stars"]
norm_reviews = fandango[num_cols]
norm_reviews.head(5)
```

```python
bar_heights= norm_reviews.loc[0,num_cols].values #  list  
bar_heights[0]= 5
print(bar_heights)
bar_postions = np.arange(5)+0.75
print(bar_postions)
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.bar( bar_postions,bar_heights,0.3)   
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(norm_reviews["Metacritic_User"],norm_reviews["IMDB"], alpha =0.5)
plt.show();
```

### demo 04 seaborn  

```python
import seaborn as sns
import matplotlib as mpl
%matplotlib inline
```

```python
def sinPlot(flip = 1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x, np.sin(x + i * 0.5) *(7-i)*flip)
```

```python
sns.set()

sinPlot()
```
