 ### demo03



```python
import random

x = range(60)
y_city = [random.uniform(15,18) for i in x]
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y_city)

plt.yticks(range(0,41,5))
x_label =["11.{} am".format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.show()
```
