 ### demo03



```python
import random

x = range(60)
y_city1 = [random.uniform(15,18) for i in x]
y_city2 = [random.uniform(1,3) for i in x]
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y_city1,color="m",linestyle="--",label="city1")
plt.plot(x,y_city2,label="city2")

plt.legend(loc="upper left")  # label

plt.yticks(range(0,41,5))
x_label =["11.{} am".format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.grid(True,linestyle="-",alpha=0.6)
plt.xlabel("Time")
plt.ylabel("Tempreature")
plt.title("demo03")
plt.show()
```

### demo04

```python
x = range(60)
y_city1 = [random.uniform(15,18) for i in x]
y_city2 = [random.uniform(1,3) for i in x]
# plt.figure(figsize=(20,8),dpi=80)
figure,axes = plt.subplots(nrows=1, ncols =2, figsize= (20,8),dpi=80)

axes[0].plot(x,y_city1,color="m",linestyle="--",label="city1")
axes[1].plot(x,y_city2,label="city2")

axes[0].legend(loc="upper left")  
axes[1].legend(loc="upper right")  



x_label =["11.{} am".format(i) for i in x]
axes[0].set_yticks(range(0,41,10))
axes[0].set_xticks(x[::10])
axes[0].set_xticklabels(x_label[::10])

axes[1].set_yticks(range(0,41,10))
axes[1].set_xticks(x[::10])
axes[1].set_xticklabels(x_label[::10])


axes[0].grid(True,linestyle="-",alpha=0.6)
axes[1].grid(True,linestyle="-",alpha=0.6)
axes[0].set_xlabel("Time")
axes[0].set_ylabel("Tempreature")

axes[1].set_xlabel("Time")
axes[1].set_ylabel("Tempreature")
plt.title("demo03")
plt.show()
```
