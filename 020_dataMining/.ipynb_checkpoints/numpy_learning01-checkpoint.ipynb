### demo09  slice in image

```python
from PIL import Image
  # iamge 
```

```python
doggy= Image.open("./doggy.jpeg")
doggy
```

```python
print(type(doggy))
```

```python
doggy_data = np.array(doggy)
doggy_data  # 3 dimension 
# doggy_data.shape
```

```python
dog1=doggy_data[0:120, ] # 第一维度： height
Image.fromarray(dog1)
```

```python
dog1=doggy_data[:,0:120 ] # 第二维度： width
Image.fromarray(dog1)
```

```python
dog1=doggy_data[::-1] # 第一维度 逆序
Image.fromarray(dog1)
dog1
```

```python
dog1=doggy_data[:,::-1]  # 第二维度 逆序
Image.fromarray(dog1)
```

```python
dog1=doggy_data[:,:,::-1]  # 第三维度 逆序
Image.fromarray(dog1)
```
